```python
import pandas as pd
import tensorflow as tf
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/kgf_dialogues.txt.txt')
df.head(5)
```

```python
df = df['powerful people create powerful empires']
```

```python
df
```

```python
movie_name = df.iloc[2:5].to_list()
```

```python
movie_name
```

```python
movie_name
```

```python
tokenizer = tf.keras.preprocessing.text.Tokenizer()
tokenizer.fit_on_texts(movie_name)
seq = tokenizer.texts_to_sequences(movie_name)
```

```python
seq[:10]
```

```python
tokenizer.word_index
```

```python
X = []
y = []
total_words_dropped = 0

for i in seq:
    if len(i) > 1:
        for index in range(1, len(i)):
            X.append(i[:index])
            y.append(i[index])
    else:
        total_words_dropped += 1

print("Total Single Words Dropped are:", total_words_dropped)
```

```python
X[:5]
```

```python
y[:8]
```

```python
X = tf.keras.preprocessing.sequence.pad_sequences(X)
```

```python
X
```

```python
X.shape
```

```python
y = tf.keras.utils.to_categorical(y)
```

```python
y
```

```python
y.shape
```

```python
vocab_size = len(tokenizer.word_index) + 1
```

```python
vocab_size
```

```python
max_sequence_len = X.shape[1]
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, 10),
    tf.keras.layers.LSTM(100, return_sequences=True),
    tf.keras.layers.LSTM(100),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(vocab_size, activation='softmax'),
])
```

```python
model.summary()
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=60)
```

```python
model.save('nwp.h5')
```

```python
def make_prediction(text, n_words):
    for i in range(n_words):
        text_tokenize = tokenizer.texts_to_sequences([text])
        text_padded = tf.keras.preprocessing.sequence.pad_sequences(text_tokenize, maxlen=14)
        prediction = np.squeeze(np.argmax(model.predict(text_padded), axis=-1))
        prediction = str(vocab_array[prediction - 1])
        print(vocab_array[np.argsort(model.predict(text_padded)) - 1].ravel()[:-3])
        text += " " + prediction
    return text
```
